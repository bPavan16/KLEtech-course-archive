```python
def pig_latin_translate(text):

  # Define vowels
  vowels = "AEIOUaeiou"

  # Check if the first letter is a vowel

  if text[0] in vowels:

      # If it is, add "way" to the end
      return text + "way"

  else:
      # If not, move the first letter to the end and add "ay"
      return text[1:] + text[0] + "ay"
```

```python
sentence = 'This is a rule based translation example'

translated_words  = [pig_latin_translate(word) for word in sentence.split() ]

print("Translated sentence : " + " ".join(translated_words))
```

# Neural Machine Translation (NMT)

```python
import torch
from torch import nn

class SimpleEncoder(nn.Module):
  def __init__(self, input_size, hidden_size):
    super(SimpleEncoder, self).__init__()

    self.rnn = nn.LSTM(input_size, hidden_size)

  def forward(self, x):
    output, (hidden, cell) = self.rnn(x)
    return hidden, cell


encoder = SimpleEncoder(input_size=10, hidden_size=20)
print("Encoder Model : " , encoder)
```

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-fr"

tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
text = "Machine translation is imporoving every day!"

inputs = tokenizer(text, return_tensors="pt", padding =True)

translated_tokens = model.generate(**inputs)

translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)

print("Original Text : " , text)
print("Translated Text : " , translated_text[0])
```

```python
class Seq2Seq(nn.Module):
  def __init__(self, encoder, decoder):
    super(Seq2Seq, self).__init__()


    self.encoder = encoder
    self.decoder=  decoder

  def forward(self ,src ,target):
    hidden ,cell = self.encoder(src)
    output = self.decoder(target, hidden, cell)
    return output


print("Seq2Seq Model is defined " )
```

```python
class AutoEncoder(nn.Module):
  def __init__(self, input_size, hidden_size):
    super(AutoEncoder, self).__init__()

    self.encoder = nn.Linear(input_size, hidden_size)
    self.decoder = nn.Linear(hidden_size, input_size)

  def forward(self, x):
    hidden = self.encoder(x)
    output = self.decoder(hidden)
    return output

autoencoder = AutoEncoder(input_size=10, hidden_size=5)

print("AutoEncoder Model = " , autoencoder)
```

# Sentence Embeddings for Translation

```python

import torch.nn as nn
import torch

class SimpleAutoencoder(nn.Module):
    def __init__(self):
        super(SimpleAutoencoder, self).__init__()
        self.encoder = nn.Linear(10, 3)  # Latent space size 3
        self.decoder = nn.Linear(3, 10)

    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return latent, reconstructed

autoencoder = SimpleAutoencoder()
input_data = torch.rand(1, 10)
latent_rep, _ = autoencoder(input_data)

print("Latent Space Representation:", latent_rep)
```

```python
class VariationalAutoencoder(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super().__init__()
        self.encoder = nn.Linear(input_dim, latent_dim)
        self.decoder = nn.Linear(latent_dim, input_dim)

    def forward(self, x):
        latent = self.encoder(x)
        reconstructed = self.decoder(latent)
        return latent, reconstructed

vae = VariationalAutoencoder(input_dim=10, latent_dim=3)
sample_data = torch.rand(1, 10)
latent_vector, _ = vae(sample_data)
print("VAE Latent Vector:", latent_vector)
```
